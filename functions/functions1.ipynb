##### Create function for  addition of two numbers only

```python
isinstance(12 ,  int)
isinstance(12 ,  (int,  float))
isinstance('12',  int)
```

```python
def  add1( x: int|float  ,  y:int|float):
    """
    This  function gives addition of two numbers
    """
    if  isinstance(x ,  (int  , float))  and \
        isinstance(y ,  (int , float)):
        z  =  x+ y
        return  z

    else:
        print('please insert  numeric data only')

```

```python
add1(2,4)
```

```python
add1('1' ,  '3')
```

```python
add1([1,2] ,  [4,5])
```

```python
add1((2,4,) ,  (3,5))
```

```python
# assert -  to raise  error
def  add2( x: int|float  ,  y:int|float):
    """
    This  function gives addition of two numbers
    """
    assert isinstance(x, (int  , float)) , "X should  be  numeric"
    assert isinstance(y  ,  (int , float)) , "Y should be numeric"
    z  =  x+ y
    return  z
```

```python
add2(2,4)
```

```python
add2(4466,-5657)
```

```python
add2('s'  ,  123)
```

```python
add2(23,  's')
```

```python
# add  eith var  length
def  add(*args):
    add  =  0
    for  i  in  args:
        add  =  add+ i
    return add    # terminate  function
```

```python
def final_bill(*args):
    bill_before_tax  = 0
    for  product  ,  price  , quantity in args:
        subtotal  =   price*quantity 
        print(f'sub total for  {product} is  {subtotal}')
        bill_before_tax =  bill_before_tax+subtotal
        
        tax  =   bill_before_tax*0.18
        bill_after_tax = tax+bill_before_tax
    return f'Bill  before tax  --> {bill_before_tax}\
    tax to be  pay  --> {tax} Bill after tax  --> {bill_after_tax}'
        
```

```python
final_bill(('laptop' , 30000 ,  2))
```

```python
def final_bill_with_discount(*args):
    bill_before_tax  = 0
    for  product  ,  price  , quantity , discount in args:
        subtotal  =   price*quantity -  price*quantity*discount/100
        print(f'sub total for  {product} is  {subtotal}')
        bill_before_tax =  bill_before_tax+subtotal
        
        tax  =   bill_before_tax*0.18
        bill_after_tax = tax+bill_before_tax
    return f'Bill  before tax  --> {bill_before_tax}\
    tax to be  pay  --> {tax} Bill after tax  --> {bill_after_tax}'
        
```

```python
final_bill_with_discount(('mobile' , 20000 , 5  ,  10))
```

### variable length  keyword arguments
**kwargs

```python
def key_args(**kwargs):
    print(type(kwargs))
    print(kwargs)
```

```python
key_args(name  =  'prajkta' ,  age  = 26 ,  education =  'masters')
```

```python
def key_args(**kwargs):
    for  key  , values in kwargs.items():
        print(f"{key} :  {values}")
```

```python
key_args(name =  'Sunita' , age  =  '23')
```

```python
key_args(gender  =  'Female' ,  Age  = 23 ,  Address  =  'pune' , education  =  'BE')
```

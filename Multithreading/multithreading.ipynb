```python
def  Hypotenus(a ,  b ):
    time.sleep(2)
    h  =  (a**2 +  b**2)**(1/2)
    return h
```

```python
# create  function for  simple interest
def simple_interest(P   ,  N  , R):
    """
    p:  principal  amount
    N :  number of  years
    R  : rate  of  ineteret
    """
    print('Simple interest  function is started')
    time.sleep(3)
    I  =  P*N*R/100
    A  =  P+I
    print(f"Amount to be return {A} ,  Interest  :{I}" )
```

```python
%%time
simple_interest(P =  10000 ,  R  =  3.5  ,  N  = 2)
```

```python
%%time
a  =  Hypotenus(11 , 3)
print(a)
simple_interest(P  =50000 ,N  = 3  , R  =  3  )
```

##### to  start both functions  at  the same  time  we can use multithreads  

```python
from  threading  import Thread
```

```python
th1  =  Thread(target  =  Hypotenus  ,  args  =  (12,  23))
th2  =  Thread(target  =  simple_interest  ,  args  =  (10000  , 2  ,  3.8))
th1.start()
th2.start()
th1.join()
th2.join()
```

```python
# create  function to find  factorial  of  number
def  fact(num):
    time.sleep(1)
    fact  =  1
    for  i in range(1 , num+1):
        fact *= i
    return fact
```

```python
fact(5)
```

```python
l  =  [ 5,6,7,8,9,10]
for  i  in l:
    print(fact(i))
```

```python
# create  function to find  factorial  of  number
def  fact(num):
    time.sleep(1)
    fact  =  1
    for  i in range(1 , num+1):
        fact *= i
    print(fact)
```

```python
thread  =  []
for i in l:
    th =  Thread(target =  fact  , args = (i , ))
    th.start()
    thread.append(th)
for  th  in  thread:
    th.join()
print(thread)
```

```python
def  multithread(l):
    thread  =  []
    for i in l:
        th =  Thread(target =  fact  , args = (i , ))
        th.start()
        thread.append(th)
    for  th  in  thread:
        th.join()
    print(thread)
```

```python
multithread([1,2,3,4,5])
```

```python
def square(n):
    print(n**2)
```

```python
# create multihtread to find  square  of n  numers  in list
```

```python
from  threading  import  Thread
```

```python
def  multithread(l:list):
    thread  =  []
    for  i in  l:
        th  =  Thread(target = square  ,  args  =  (i , ))
        th.start()
        thread.append(th)
    for  th  in  thread:
        th.join()
    print(thread)
```

```python
multithread([23,45,53,6,-3,-2])
```

### download  file  from  url

```python
from  urllib.request import  urlretrieve
```

```python
url  =  'https://raw.githubusercontent.com/utkarshg1/mlproject_regression/main/artifacts/data.csv'
```

```python
url.split('/')[-1]
```

```python
def download(url1):
   file_name = url1.split('/')[-1]
   print(f"{file_name}  is started  to  download")
   urlretrieve(url1 , file_name )
   print(f"{file_name}  downloaded")
```

```python
download(url1= url)
```

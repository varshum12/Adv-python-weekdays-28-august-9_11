```python
class Person:
    def __init__(self , Name  , Age):
        self.Name =Name
        self.Age  = Age

# define  info method
    def  info(self):
        print(f'I am {self.Name}')
        print(f'I am {self.Age}  years  old')


# create  child class  Employee 
class  Employee(Person):
    def __init__(self , Name  , Age , Company):
        super().__init__(Name  , Age )
        self.Company  =  Company


# def  work method
    def work(self):
        super().info()      # self  u can  self also
        return  f"I work at  {self.Company}"                               
```

#### Multilevel inheritance
- class  Employee  , Manager  , Project_manager

```python
class Employee:
    def __init__(self , emp_name  , emp_id):
        self.emp_name  =  emp_name
        self.emp_id  =  emp_id

    def info(self):
        print(f'Name  of Employee is  {self.emp_name}')
        print(f'Id  of employee is  {self.emp_id}')


# add child class manager
class  Manager(Employee):
    def __init__(self  , emp_name  , emp_id  ,  Department):
        super().__init__(emp_name  , emp_id)
        self.Department  = Department

    # add method manager_info
    def Manager_info(self):
        super().info()
        return  f'Department : {self.Department}'


## add Grand  child class project manager
class Project_Manager(Manager):
    def __init__(self,  emp_name  , emp_id  , Department  , Project):
        super().__init__(emp_name  , emp_id  , Department )
        self.Project  = Project


## add  method  project info
    def  Project_info(self):
        return  f"{self.Manager_info()} ,  \
           Project :{self.Project}   "

```

```python
E1  =  Employee(emp_name  = 'Suresh' ,  emp_id  = 1)
```

```python
E1.emp_name
E1.emp_id
E1.info()
```

```python
M1  =  Manager(emp_name  =  'sarthak' ,  emp_id =  2  ,  Department = 'Engineering')
```

```python
M1.info()
```

```python
M1.Manager_info()
```

```python
M1.Project_info()
```

#### Create class with  use  , admin  and  regular user

```python
class User:
    def  __init__(self  ,  email ,  password ):
        self.email  = email 
        self.password  = password

    def  User_info(self):   
        return f"email : {self.email} ,  password  :  {self.password} "
    


# create  class  admin
class Admin(User):
    def __init__(self ,  email ,  password  ,  Access):
        super().__init__(email  ,  password)
        self.Access =  Access

    def  Admin_info(self):
        return f"{self.User_info()} ,  Access : {self.Access}"
    

# regular User class

class Regular_User(User):
    def __init__(self  , email , password  , Subscription):
        super().__init__(email  ,  password)
        self.Subscription =  Subscription

    def  Regular_user_info(self):
        return  f"{self.User_info()} ,  Subscription  :  {self.Subscription}"
    
        
                   
```

## Data validation for  student class
name  : str
gender 
roll_no  : int
Marks :  float
email  : EmailStr
Hobbies :  List[str]

```python
from  pydantic  import BaseModel ,  EmailStr  , Field
from  typing import Literal ,  List 
```

```python
class  Student(BaseModel):
        name  : str  =  Field(description  =  "Enter name here")   
        gender  :  Literal['Male' ,  'Female']
        roll_no  : int =  Field( description  =  "Enter  roll no here"  , gt =  0  )
        Marks  : float = Field(description  =  "Enter   Marks here"  , gt =  0  )  
        email :  EmailStr =  Field(description  =  'entere  mail  here')
        Hobbies :  List[str] =  Field(description  =  'enter hobbies here'  ,  default_factory= List ,  )
```

```python
S1  =  Student(name  =  123  ,  roll_no  =  21  ,  Marks  =  45 , gender  =   "Male" , email  =  'example@gmail.com'  , Hobbies  = ['Reading']   )
```
